```python
import unittest
import sys

class Customer:
    no_of_vehicles = 0
    hours = 0
    days = 0
    weeks = 0
    def __init__(self,no_of_vehicles,hours,days,weeks):
        self.no_of_vehicles = no_of_vehicles
        self.hours = hours
        self.days = days
        self.weeks = weeks

class RentalServices:
    vehicles_available = 50
    amount = 0
    def __init__(self,vehicles_available):
        self.vehicles_available = vehicles_available
    
    def calculateBill(self,c):
        amount = c.no_of_vehicles*((c.hours*5)+(c.days*20)+(c.weeks)*50)
        return amount

    def applyDiscount(self,c):
        amount = (self.calculateBill(c))*0.7
        return amount

    def displayBill(self,c):
        print("************** INVOICE **************")
        print("Number of vehicles ordered: ",c.no_of_vehicles)
        print("Time spent: ",c.weeks,"weeks,",c.days,"days,",c.hours,"hours.")
        amount = self.calculateBill(c)
        if c.no_of_vehicles <=2:
            print("Total Amount: ", amount)
        else:
            print("Total Amount: ", amount)
            print("Discount  applied: 30%")
            print("Total amount after applying discount: ", self.applyDiscount(c))
            print("*********** THANK YOU **************")

class TestLibrary(unittest.TestCase):

    def test_calculateBill(self):
        c = Customer(no_of_vehicles = 2,hours = 2,days = 4,weeks = 2)
        r = RentalServices(50)
        self.assertEqual(r.calculateBill(c), 380)
    
    def test_applyDiscount(self):
        c = Customer(no_of_vehicles = 5,hours = 2,days = 4,weeks = 3)
        r = RentalServices(50)
        self.assertEqual(r.applyDiscount(c), 840)

def main():
    vehicles_available = int(input("Enter the no.of vehicles available in shop: "))
    
    while True:
        r = RentalServices(vehicles_available)
        print("**************** Welcome to Car Rental Services ****************")
        print(r.vehicles_available," vechicles available")
        no_of_vehicles = int(input("Please enter no.of vehicles needed: "))
        if no_of_vehicles > vehicles_available:
            print("Sorry! only",vehicles_available," vehicles are available.")
            cont = input("Do you want to rent more vehicles? Y/N : ")
            if cont == "Y" or cont == "y":
                continue
            if cont == "N" or "n":
                print("**************Thank you for using our services****************\n")
                break
        print("Please enter the time duration")
        weeks = int(input("no.of weeks: "))
        days = int(input("no.of days: "))
        if days >= 7:
            print("Error: Days must be less than 7. If more than 7, add to weeks.\n")
            continue
        hours = int(input("no.of hours: "))
        if hours > 24:
            print("Error: Hours must be less than 24. If more than 24, add to days.\n")
            continue
        c = Customer(no_of_vehicles=no_of_vehicles,weeks=weeks,days=days,hours=hours)
        r.displayBill(c)
        vehicles_available = vehicles_available - no_of_vehicles
        cont = input("Do you want to rent more vehicles? Y/N : ")
        if cont == "Y" or cont == "y":
            pass
        elif cont == "N" or "n":
            print("**************Thank you for using our services****************\n")
            sys.exit()
            
            


if __name__ == "__main__":
    main()
    unittest.main()
```
